## Read In Data

```python
DATADIR = '/Users/admin/notebooks/fdic_data/financial_data_raw/'
FDIC_CODES = [3510, 3511, 33238, 7213, 628]
```

```python
data_files = os.listdir(DATADIR)
```

```python
dicts = list()

for item in data_files:
    if item not in ('.ipynb_checkpoints', 'mdrm_codes'):
        filepath = os.path.join(DATADIR, item)
        print(filepath)
        readin = pd.read_csv(filepath, header=[0, 1], sep='\t')
        readin.columns = readin.columns.droplevel(1)
        filtered = readin[readin['FDIC Certificate Number'].isin(FDIC_CODES)]
        dicts.append(filtered)
```

## Review

```python
df = pd.concat(dicts)
```
